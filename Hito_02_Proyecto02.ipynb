```python
from scipy.stats import mannwhitneyu
import numpy as np
import pandas as pd
from IPython.display import HTML, display

df = client.list_rows(table).to_dataframe()

alpha= 0.05

#BPM
# Almacenar streams en listas por categoría
streams_alto_bpm = []
streams_bajo_bpm = []

# Iterar sobre las filas del DataFrame
for index, row in df.iterrows():
    streams = row['streams_limpio']
    # Usar comparación correcta para 'cat_fin_bpm'
    if row['cat_fin_bpm'] == 'Alto':
        streams_alto_bpm.append(streams)
    elif row['cat_fin_bpm'] == 'Bajo':
        streams_bajo_bpm.append(streams)

# Calcular promedios por categoría
avg_streams_alto = np.mean(streams_alto_bpm)
avg_streams_bajo = np.mean(streams_bajo_bpm)

# Aplicar el test de Wilcoxon (Mann-Whitney U test) a cada característica
statistic, p_value = mannwhitneyu(streams_alto_bpm, streams_bajo_bpm, alternative='two-sided')
print( "BPM")
print("Estadística U:", statistic)
print("Valor p :", p_value)
print("Análisis para la característica:")
if p_value < alpha: print("Se rechaza la hipótesis nula: Existe una diferencia significativa entre las categorías 'alto' y 'bajo'")
else: print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
print()

#ACOUSTICNESS

streams_alto_acousticness = []
streams_bajo_acousticness = []
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_acousticness'] == 'Alto':
        streams_alto_acousticness.append(streams)
    elif row['cat_fin_acousticness'] == 'Bajo':
        streams_bajo_acousticness.append(streams)
statistic, p_value = mannwhitneyu(streams_alto_acousticness, streams_bajo_acousticness, alternative='two-sided')
print( "ACOUSTICNESS")
print("Estadística U:", statistic)
print("Valor p :", p_value)
print("Análisis para la característica:")
if p_value < alpha: print("Se rechaza la hipótesis nula: Existe una diferencia significativa entre las categorías 'alto' y 'bajo'")
else: print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
print()

#DANCEABILLITY

streams_alto_danceability = []
streams_bajo_danceability = []
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_danceability'] == 'Alto':
        streams_alto_danceability.append(streams)
    elif row['cat_fin_danceability'] == 'Bajo':
        streams_bajo_danceability.append(streams)
statistic, p_value = mannwhitneyu(streams_alto_danceability, streams_bajo_danceability, alternative='two-sided')
print( "DANCEABILLITY")
print("Estadística U:", statistic)
print("Valor p :", p_value)
print("Análisis para la característica:")
if p_value < alpha: print("Se rechaza la hipótesis nula: Existe una diferencia significativa entre las categorías 'alto' y 'bajo'")
else: print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
print()

#ENERGY

streams_alto_energy = []
streams_bajo_energy = []
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_energy'] == 'Alto':
        streams_alto_energy.append(streams)
    elif row['cat_fin_energy'] == 'Bajo':
        streams_bajo_energy.append(streams)
statistic, p_value = mannwhitneyu(streams_alto_energy, streams_bajo_energy, alternative='two-sided')
print( "#ENERGY")
print("Estadística U:", statistic)
print("Valor p :", p_value)
print("Análisis para la característica:")
if p_value < alpha: print("Se rechaza la hipótesis nula: Existe una diferencia significativa entre las categorías 'alto' y 'bajo'")
else: print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
print()

#INSTRUMENTALNESS

streams_alto_instrumentalness = []
streams_bajo_instrumentalness = []
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_instrumentalness'] == 'Alto':
        streams_alto_instrumentalness.append(streams)
    elif row['cat_fin_instrumentalness'] == 'Bajo':
        streams_bajo_instrumentalness.append(streams)
statistic, p_value = mannwhitneyu(streams_alto_instrumentalness, streams_bajo_instrumentalness, alternative='two-sided')
print( "INSTRUMENTALNESS")
print("Estadística U:", statistic)
print("Valor p :", p_value)
print("Análisis para la característica:")
if p_value < alpha: print("Se rechaza la hipótesis nula: Existe una diferencia significativa entre las categorías 'alto' y 'bajo'")
else: print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
print()

#LIVENESS

streams_alto_liveness = []
streams_bajo_liveness = []
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_liveness'] == 'Alto':
        streams_alto_liveness.append(streams)
    elif row['cat_fin_liveness'] == 'Bajo':
        streams_bajo_liveness.append(streams)
statistic, p_value = mannwhitneyu(streams_alto_liveness, streams_bajo_liveness, alternative='two-sided')
print( "LIVENESS")
print("Estadística U:", statistic)
print("Valor p :", p_value)
print("Análisis para la característica:")
if p_value < alpha: print("Se rechaza la hipótesis nula: Existe una diferencia significativa entre las categorías 'alto' y 'bajo'")
else: print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
print()

#SPEECHINESS

streams_alto_speechiness = []
streams_bajo_speechiness = []
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_speechiness'] == 'Alto':
        streams_alto_speechiness.append(streams)
    elif row['cat_fin_speechiness'] == 'Bajo':
        streams_bajo_speechiness.append(streams)

# Verificar y eliminar valores NaN
streams_alto_speechiness = [x for x in streams_alto_speechiness if pd.notna(x)]
streams_bajo_speechiness = [x for x in streams_bajo_speechiness if pd.notna(x)]

# Verificar que ambos conjuntos de datos tengan algún tamaño
if len(streams_alto_speechiness) > 0 and len(streams_bajo_speechiness) > 0:
  statistic, p_value = mannwhitneyu(streams_alto_speechiness, streams_bajo_speechiness, alternative='two-sided')
  print( "SPEECHINESS")
  print("Estadística U:", statistic)
  print("Valor p :", p_value)
else:
    print("No hay datos válidos para realizar la prueba con la variable Speechiness.")

#VALENCE

streams_alto_valence = []
streams_bajo_valence = []
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_valence'] == 'Alto':
        streams_alto_valence.append(streams)
    elif row['cat_fin_valence'] == 'Bajo':
        streams_bajo_valence.append(streams)
statistic, p_value = mannwhitneyu(streams_alto_valence, streams_bajo_valence, alternative='two-sided')
print( "VALENCE")
print("Estadística U:", statistic)
print("Valor p :", p_value)
print("Análisis para la característica:")
if p_value < alpha: print("Se rechaza la hipótesis nula: Existe una diferencia significativa entre las categorías 'alto' y 'bajo'")
else:
  print("No se puede rechazar la hipótesis nula: <b>No hay una diferencia significativa entre las categorías 'alto' y 'bajo'*</b>")
print()

```
